```python
import json
import itertools
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.efficientnet import preprocess_input

# -----------------------------
# CONFIG (Windows paths)
# -----------------------------
BASE_DIR   = Path(r"C:\Users\sagni\Downloads\MRI Scan")
TRAIN_DIR  = Path(r"C:\Users\sagni\Downloads\MRI Scan\archive\Training")
TEST_DIR   = Path(r"C:\Users\sagni\Downloads\MRI Scan\archive\Testing")

MODEL_H5   = BASE_DIR / "model.h5"
CLASS_PKL  = BASE_DIR / "class_indices.pkl"
METRICS_JSON = BASE_DIR / "metrics.json"
CR_CSV     = BASE_DIR / "classification_report.csv"
CM_CSV     = BASE_DIR / "confusion_matrix.csv"

ACC_PNG    = BASE_DIR / "accuracy_loss_view.png"
CM_PNG     = BASE_DIR / "confusion_matrix_view.png"

# Must match training
IMG_SIZE   = (256, 256)
BATCH_SIZE = 16

# -----------------------------
# Utils
# -----------------------------
def show_accuracy_loss(metrics_json: Path):
    """Show and save accuracy/loss curves from metrics.json."""
    if not metrics_json.exists():
        print(f"[WARN] {metrics_json} not found. Skipping accuracy/loss plot.")
        return

    with open(metrics_json, "r", encoding="utf-8") as f:
        m = json.load(f)

    hist = m.get("history", {})
    acc, val_acc = hist.get("accuracy", []), hist.get("val_accuracy", [])
    loss, val_loss = hist.get("loss", []), hist.get("val_loss", [])
    if not (acc and val_acc and loss and val_loss):
        print("[WARN] metrics.json has no history arrays. Skipping accuracy/loss plot.")
        return

    epochs = range(1, len(acc) + 1)
    plt.figure(figsize=(10, 8))

    # Accuracy
    plt.subplot(2, 1, 1)
    plt.plot(epochs, acc, label="Train Acc")
    plt.plot(epochs, val_acc, label="Val Acc")
    plt.title("Training & Validation Accuracy")
    plt.xlabel("Epoch"); plt.ylabel("Accuracy")
    plt.grid(alpha=0.3); plt.legend()

    # Loss
    plt.subplot(2, 1, 2)
    plt.plot(epochs, loss, label="Train Loss")
    plt.plot(epochs, val_loss, label="Val Loss")
    plt.title("Training & Validation Loss")
    plt.xlabel("Epoch"); plt.ylabel("Loss")
    plt.grid(alpha=0.3); plt.legend()

    plt.tight_layout()
    plt.savefig(ACC_PNG, dpi=200)
    print(f"[INFO] Saved accuracy/loss plot â†’ {ACC_PNG}")
    plt.show()

def recompute_confusion_matrix():
    """If CM CSV is missing, compute it from model + Testing set and save CSVs."""
    import pickle

    if not MODEL_H5.exists():
        raise FileNotFoundError(f"Missing model: {MODEL_H5}")
    if not CLASS_PKL.exists():
        raise FileNotFoundError(f"Missing class map: {CLASS_PKL}")
    if not TEST_DIR.exists():
        raise FileNotFoundError(f"Missing Testing dir: {TEST_DIR}")

    print("[INFO] Recomputing confusion matrix from test setâ€¦")

    with open(CLASS_PKL, "rb") as f:
        class_indices = pickle.load(f)
    idx_to_class = {v: k for k, v in class_indices.items()}
    classes = [idx_to_class[i] for i in range(len(idx_to_class))]
    print("[INFO] Classes:", classes)

    test_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
    test_flow = test_gen.flow_from_directory(
        str(TEST_DIR),
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        classes=classes,
        class_mode="categorical",
        shuffle=False,
    )

    model = load_model(str(MODEL_H5))
    probs = model.predict(test_flow, verbose=1)
    y_pred = np.argmax(probs, axis=1)
    y_true = test_flow.classes

    # Save CSVs for reuse
    cm = confusion_matrix(y_true, y_pred, labels=list(range(len(classes))))
    pd.DataFrame(cm, index=classes, columns=classes).to_csv(CM_CSV)
    cr = classification_report(y_true, y_pred, target_names=classes, output_dict=True, zero_division=0)
    pd.DataFrame(cr).to_csv(CR_CSV)
    print(f"[INFO] Saved confusion_matrix.csv â†’ {CM_CSV}")
    print(f"[INFO] Saved classification_report.csv â†’ {CR_CSV}")

def show_confusion_matrix(cm_csv: Path):
    """Show and save normalized confusion-matrix heatmap from CSV."""
    if not cm_csv.exists():
        recompute_confusion_matrix()

    cm_df = pd.read_csv(cm_csv, index_col=0)
    cm = cm_df.values
    labels = cm_df.index.tolist()

    # Normalize row-wise
    cm_norm = cm.astype("float") / cm.sum(axis=1, keepdims=True)
    cm_norm = np.nan_to_num(cm_norm)

    fig = plt.figure(figsize=(9, 7))
    ax = plt.gca()
    im = ax.imshow(cm_norm, interpolation="nearest", cmap="viridis")
    plt.title("Confusion Matrix (Normalized)")
    cbar = plt.colorbar(im, fraction=0.046, pad=0.04)
    cbar.ax.set_ylabel("Proportion", rotation=90)

    ticks = np.arange(len(labels))
    plt.xticks(ticks, labels, rotation=45, ha="right")
    plt.yticks(ticks, labels)

    thresh = cm_norm.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        ax.text(j, i, f"{cm[i,j]}\n{cm_norm[i,j]*100:.1f}%",
                ha="center", va="center",
                color="white" if cm_norm[i, j] > thresh else "black",
                fontsize=9)

    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.tight_layout()
    plt.savefig(CM_PNG, dpi=220)
    print(f"[INFO] Saved confusion matrix heatmap â†’ {CM_PNG}")
    plt.show()

# -----------------------------
# Run
# -----------------------------
if __name__ == "__main__":
    BASE_DIR.mkdir(parents=True, exist_ok=True)

    # 1) Accuracy & Loss
    show_accuracy_loss(METRICS_JSON)

    # 2) Confusion Matrix
    show_confusion_matrix(CM_CSV)
```
